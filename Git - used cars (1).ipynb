## 1. Type of Gear-Box is distributed

```python
# set colors for chart
colors = ['#DC143C', '#fa9b98', '#0d8c68', '#95c5b1', '#4169E1', '#adc3ff', '#595959', '#c4c4c4']
x = bmw.transmission.value_counts()

# set size for chart
fig, ax = plt.subplots (figsize = (10,6))

# set chart size and colors
ax.barh ([0], x.values[0], height = 0.7, color = colors[5], alpha = 0.7)

# text
plt.text (-1400, 0, 'Gear-box type: Semi-Auto',  {'font':'Serif', 'weight':'bold', 'Size':'12', 'style':'normal', 'color':'black'} , alpha = 0.7)

plt.text(2500,0, '43%',{'font':'Serif', 'weight':'bold','size':'12','color':'black'}, alpha = 0.7)

# set chart size and colors
ax.barh([1],x.values[1],height = 0.7, color = '#b20710', alpha = 0.7)

plt.text(-1400,1, 'Gear-box type: Automatic', {'font': 'Serif','weight':'bold','Size': '12','style':'normal', 'color':'black'}, alpha = 0.7)

plt.text(2000,1, '33%',{'font':'Serif', 'weight':'bold','size':'12','color':'black'}, alpha = 0.7)

ax.barh([2],x.values[2],height = 0.7, color = colors[2], alpha = 0.7)
plt.text(-1400,2, 'Gear-box type: Manual', {'font': 'Serif','weight':'bold','Size': '12','style':'normal', 'color':'black'}, alpha = 0.7)
plt.text(1300,2, '24%',{'font':'Serif', 'weight':'bold','size':'12','color':'black'}, alpha = 0.7)
 
# setup 
fig.patch.set_facecolor('#f6f5f5')
ax.set_facecolor('#f6f5f5')


plt.text(-1150,3, 'How Type of Gear-Box Is Distributed? ' ,{'font': 'Serif', 'Size': '25','weight':'bold', 'color':'black'}, alpha = 0.9)
ax.axes.get_xaxis().set_visible(False)
ax.axes.get_yaxis().set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(True)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
fig.show()
```

## 2.Type of Engineer Fuel

```python
bmw.fuelType.value_counts()
```

```python
fig, ax = plt.subplots (figsize = (10,7), dpi = 90)
ax.set_facecolor ('#f6f5f5')
fig.patch.set_facecolor('#f6f5f5')

# Values
a =bmw.fuelType.value_counts()
ax.bar ( x= a.index , height = a.values, color = 'gray', alpha = 0.5)

# Labels on the axes
label = a.values.tolist()

# Parameters per column
for i in range(len(label)):
    x = i  - 0.35
    y = (i+18)/2 + label[i]
    x = x +0.25 # position the X axis
    y = y + 52 # position the Y axis
    ax.text(x,y, '{}'.format(a.values[i]),{'font': 'serif', 'weight': 'normal', 'color': 'black', 'fontsize': 10}, alpha = 0.8)
    
    
    # Text
    ax.text(-1,9000, 'Fuel Type | Type of Engine Fuel', {'font': 'serif', 'color': 'black', 'fontsize': 18, 'weight':'bold'},alpha = 0.9 )
ax.text(-1,8000, 'Well it seems like most number of cars have Diesel Engine. ', {'font': 'serif', 'color': 'black', 'fontsize': 12,},alpha = 0.8)

# Delete axes

for loc in ['left','right','top','bottom']:
    ax.spines[loc].set_visible(False)
    
# Clear axes unit values
ax.axes.get_yaxis().set_visible(False)
```

## 3.Distribution of Model

```python

fig, ax = plt.subplots(figsize = (15,6))
ax.set_facecolor('#f6f5f5')
fig.patch.set_facecolor('#f6f5f5')

a = bmw.model.value_counts()
colors = ['#b20710','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey']
ax.bar(x = a.index, height = a.values, color = colors, alpha = 0.9)

# Create labels
label = a.values.tolist()
 
# Text on the top of each bar
for i in range(len(label)):
    x = i  - 0.35
    y = (i+18)/2 + label[i]
    x = x-0.08
    y = y + 52
    ax.text(x,y, '{}'.format(a.values[i]),{'font': 'serif', 'weight': 'normal', 'color': 'black', 'fontsize': 10}, alpha = 0.8)

#ax.set_xticks(a.values)
ax.text(-1,2800, 'Distribution of Model', {'font': 'serif', 'color': 'black', 'fontsize': 18, 'weight':'bold'},alpha = 0.9 )
ax.text(-1,2700, ' Most of the cars are of have model 3 Series . ', {'font': 'serif', 'color': 'black', 'fontsize': 12,},alpha = 0.8)

for loc in ['left','right','top','bottom']:
    ax.spines[loc].set_visible(False)
    
ax.set_xticklabels(a.index, {'font': 'serif', 'color': 'black', 'fontsize': 8,'weight':'bold'}, alpha = 0.8)
ax.axes.get_yaxis().set_visible(False)
fig.show()
```
